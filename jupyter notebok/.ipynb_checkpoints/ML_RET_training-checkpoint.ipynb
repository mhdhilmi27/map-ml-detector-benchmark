```python
import os
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
training_path = '/Users/hilmi/Documents/ML_RET/TiltIT/batam/ta_profile/data_training'
```

```python
first_file = 0
for file in os.listdir(training_path):
    if first_file == 0 and '.csv' in file:
        df = pd.read_csv('%s/%s' % (training_path, file), encoding='ISO-8859-1')
        print('%s : %s' % (file, len(df)))
        first_file = 1
    elif first_file == 1 and '.csv' in file:
        df_temp = pd.read_csv('%s/%s' % (training_path, file), encoding='ISO-8859-1')
        print('%s : %s' % (file, len(df_temp)))
        df = pd.concat([df, df_temp], axis=0, ignore_index=True, sort=False)

print('             Total: %s' % len(df))
```

```python
df = df[['eutrancellfddid','isd','height','perc55_ta_distance_km','perc65_ta_distance_km','perc75_ta_distance_km','main_distance_km','coverage_issue','current_tilt','proposed_tilt']]
```

```python
df['coverage_issue'].unique()
```

```python
df.loc[(pd.isnull(df['coverage_issue'])), 'coverage_issue'] = 0
df.loc[(df['coverage_issue'] == 'Normal'), "coverage_issue"] = 0
df.loc[(df['coverage_issue'] == 'Overshoot_half_isd_2km'), "coverage_issue"] = 1
df.loc[(df['coverage_issue'] == 'Overshoot_half_isd_1km'), "coverage_issue"] = 1
df.loc[(df['coverage_issue'] == 'Overshoot_half_isd'), "coverage_issue"] = 1
df.loc[(df['coverage_issue'] == 'Overshoot_one_isd'), "coverage_issue"] = 1
df.loc[(df['coverage_issue'] == 'Undershoot'), "coverage_issue"] = -1
```

# ML

```python
features = ['isd', 'height', 'perc55_ta_distance_km', 'perc65_ta_distance_km',
            'perc75_ta_distance_km', 'main_distance_km', 'coverage_issue', 'current_tilt']
```

```python
target = 'proposed_tilt'
```

```python
# Drop rows with missing values to ensure clean input
df = df.dropna(subset=features + [target])
```

```python
# Prepare features and labels
X = df[features]
y = df[target]
cellname = df['eutrancellfddid']
```

```python
# Train-test split while preserving index and cellname
X_train, X_test, y_train, y_test, cellname_train, cellname_test = train_test_split(
    X, y, cellname, test_size=0.2, random_state=42
)

# Ensure all features are numeric
X_train['coverage_issue'] = X_train['coverage_issue'].astype(int)
X_test['coverage_issue'] = X_test['coverage_issue'].astype(int)

# Train XGBoost model
xgb_model = XGBRegressor(n_estimators=100, random_state=42, objective='reg:squarederror')
xgb_model.fit(X_train, y_train)

# Predict
y_pred = xgb_model.predict(X_test)

# Evaluate MAE
mae = mean_absolute_error(y_test, y_pred)
```

```python
# Assemble result DataFrame
result_df = X_test.copy()
result_df['eutrancellfddid'] = cellname_test.values
result_df['proposed_tilt_actual'] = y_test.values
result_df['proposed_tilt_predicted'] = y_pred
result_df = result_df.sort_values(by='eutrancellfddid')
```

```python
import matplotlib.pyplot as plt

importances = xgb_model.feature_importances_
features = X_train.columns

importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
plt.barh(importance_df['Feature'], importance_df['Importance'])
plt.xlabel('Importance Score')
plt.title('XGBoost Feature Importances')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
